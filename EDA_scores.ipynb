```python
import pandas as pd
url = "https://tinyurl.com/5n96xpkd"
df = pd.read_csv(url)
df
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
x=df['Hours'].values.reshape(-1,1)
y=df['Scores'].values.reshape(-1,1)
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.xlabel('Number of Hours Studied')
plt.ylabel('Scores')
plt.title('Hours vs. Score')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
predictions=model.predict(x)

plt.plot(x,predictions,c='black',linewidth=1)
plt.scatter(x,y)
plt.xlabel('Number of Hours Studied')
plt.ylabel('Scores')
plt.title('Hours vs. Score')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
import numpy as np
prediction = np.array([[9.25]])
predicted_score = model.predict(prediction)
predicted_score_int = int(predicted_score)
print("Predicted Score for 9.25 Hours of Study:", predicted_score_int)
```
